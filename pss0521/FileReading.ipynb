# Traverse a directory
* 如何剖析一個結構化的文字檔，例如.csv。.csv的檔案格式和.txt是一模一樣的，都是plain text，唯一的差別在於.csv檔全部都用逗號分格，還有副檔名不一樣而已。若把副檔名都改為.txt基本上還是一樣的。

```python
jydict = {}
for fname in os.listdir(datapath):
    year = int(fname[-8:-4])
    fin = open(datapath + fname, 'r')
    fstring = fin.read()
    lines = fstring.split('\n')
    for line in lines[3:-1]:
        cells = line.split(';')
        jydict.setdefault(cells[0], {})
        if len(cells) >=8:
            if len(cells[3]) == 0:
                IF = 0
            else:
                IF = float(cells[3])
            jydict[cells[0]].setdefault(year, IF)
for j in jydict:
    for y in jydict[j]:
        print j, y
```

```python
import xlrd
outlist = []
fname = 'data/JCR_Communication.xlsx'
sheetname = "jcr_c2012"
bk = xlrd.open_workbook(fname)
try:
    sh = bk.sheet_by_name(sheetname)
except:
    print "no sheet in %s named data" % fname
print "nrows %d, ncols %d" % (sh.nrows, sh.ncols)
if sh.ncols==1:
    for rownum in range(sh.nrows):
        outlist.append(sh.cell_value(rownum, 0))
else:
    for rownum in range(sh.nrows):
        temp = []
        for colnum in range(sh.ncols):
            temp.append(sh.cell_value(rownum, colnum))
        outlist.append(temp)
```
