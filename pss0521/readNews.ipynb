```python
import os
import re
datapath = 'data/libnew/'
filelist = []
start_reg_exp = re.compile('(.*)[.](.*) [|] (\d{4}[-/]\d{2}[-/]\d{2})')
'''
newsData = []
對每一行：
如果 發現有「230. 自由時報 | 2010-01-31」的pattern
    我就知道新的新聞進來了，產生一個news ＝{}
    跳三行到下四行
不然
    如果發現發現「文章編號: [201001312460117]」的pattern
        就把news dictionary存到newsData [] 
    不然
        就一直收新聞的content，直到
    往下一行
'''


for fname in os.listdir(datapath):
    if fname.endswith('.txt'):
        filelist.append(datapath + fname)
newsData = []
for f in filelist:
    print f
    lines = []
    fin = open(f, 'r')
    for line in fin:
        if len(line) > 5:
            lines.append(line)
    i = 0
    while i < len(lines): 
        matches_list=start_reg_exp.findall(lines[i])
        if len(matches_list) != 0: 
            news = {}
            news['date'] = matches_list[0][2]
            news['title'] = lines[i+2]
            news['agency'] = matches_list[0][1]
            news['content'] = ""
            news['tags'] = lines[i+1]
#             print matches_list[0][2], '\t', lines[i+2]
            i += 4
        else:
            matches_end=re.findall('文章編號.*\[(\d{15})\]', lines[i])
            if len(matches_end) > 0:
                news['id'] = matches_end[0]
#             if '文章編號' in lines[i]:
                newsData.append(news)
                i += 2
            else:
                news['content'] += lines[i]
            i += 1
        
```

```python
import os
import re
from time import strptime
datapath = 'data/libnew/'
filelist = []
start_reg_exp = re.compile('(.*)[.](.*) [|] (\d{4}[-/]\d{2}[-/]\d{2})')

for fname in os.listdir(datapath):
    if fname.endswith('.txt'):
        filelist.append(datapath + fname)
newsData = []
for f in filelist:
    print f
    lines = []
    fin = open(f, 'r')
    for line in fin:
        if len(line) > 5:
            lines.append(line)
#     i = 0
#     for line in lines:
#     for i in range(len(lines)):
    i = 0
    level = 0
    while i < len(lines): 
        matches_list=start_reg_exp.findall(lines[i])
        if len(matches_list) != 0:
            news = {}
            news['head'] = lines[i+1]
            news['title'] = lines[i+2]
            news['date'] = strptime(matches_list[0][2], '%Y-%m-%d')
            news['content'] = ""
#             print matches_list
#             print matches_list[0][2], '\t', lines[i+2]
            i += 4
            level = 1
        elif level == 1:
            matches_end=re.findall('文章編號.*\[(\d{15})\]', lines[i])
            if len(matches_end) != 0:
#                 print matches_end
                news['id'] = matches_end[0]
                newsData.append(news)
                level = 0
            else:
                news['content']+=lines[i]
            i += 1
        else:
            i += 1
        
```
