# 9X9 Table
* 以下是一個3/19所上過的9X9乘法表的範例，目的是讓你了解for-each作為一個模組被包含在另一個for-each裡面的情形。有助於了解for-each的使用概念。

```python
for i in range(1, 10):
    for j in range(1, 10):
        print "%d*%d=%2d"%(i, j, i*j),
    print
```
