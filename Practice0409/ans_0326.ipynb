1.  事實上上面PM2.5是濃度的平均數，但是，通常在報導的時候會對應到一個表格，分別標示現在PM2.5濃度對應到的指標為多少（1~10），並據此分為低（L）、中（M）、高（H）、非常高（vH）四個等級。請利用list、dictionary、或者是if-else，想辦法把上述的濃度所對應到的指標（1~10）與等級（L, M, H, vH）列印出來。
    * 對應連結如下：http://taqm.epa.gov.tw/taqm/tw/fpmi.aspx
2. 你會遇到這個問題，因為這些資料有些站會沒有資料，你覺得這時候可以怎麼處理？
3. 最好的狀況是你可以印出下面這個樣子。

```python
maplist = [-1, 11, 23, 35, 41, 47, 53, 58, 64, 70, 100]
leveldict = {-1: 'unknown', 1:'L', 2:'L', 3:'L', 4:'M', 5:'M', 6:'M', 7:'H', 8:'H', 9:'H', 10:'vH'}
results = []
for site in aqx:
    index = -1
    for i in range(len(maplist)-1):
        if len(site['PM2.5']) > 0:
            if maplist[i] < int(site['PM2.5']) <=maplist[i+1]:
                index = i+1
        else:
            index = -1
            site['PM2.5'] = "-1"
    if leveldict.has_key(index):
        level = leveldict[index]
    else:
        level = None
    results.append([int(site['PM2.5']), index, level, site['SiteName'], site['County']])
for k, i, l, s, c in sorted(results, reverse=True):
    print k, i, l, s, c
            
```
