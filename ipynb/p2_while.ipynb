## Scenario
* 有時候，我們並不知道究竟有多少資料筆數，在這樣的狀況下，沒辦法用for-each列數完所有的資料。
* 比方說，下面這筆新聞資料。首先，程式可以判斷究竟有幾行，而大約透過日期或者文章編號我就可以知道究竟有幾筆資料。然而，我卻不知道，每一篇新聞的內容有幾行，所以當如果我要把新聞內容用一個變數存放的時候，我必須想個辦法「持續的」讀新聞進來，直到讀到「文章編號」的那一行為止。
* 在這種狀況下就需要用一個語法叫做while。

## What is while
* 先想想英文的while是什麼意思，和if有何不同。
    * if指的是，「如果」滿足某一個條件，那就做這件事。
    * 但是while也有「當」滿足某個條件，或者有些人講得太白話，就會變成「如果」滿足某個條件...這樣不就和if很像？
    * 但嚴格上來說while的用法和if在句子中的用法是有所不同的。while有持續進行的意思，e.g., while you are sleeping...
* 所以要用while來解決上述的問題，語法大致上是while(not read 文章編號) do something
    * 用程式邏輯來說，叫做while line has no "文章編號"
    * 更接近程式邏輯一點 while "文章編號" not in line:
