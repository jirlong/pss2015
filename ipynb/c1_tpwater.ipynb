## Data Field
0. update_date 日期 
1. update_time : 時間 
2. qua_id : 監控站ID 
3. code_name : 監控站名稱 
4. longitude : 經度 
5. latitude: 緯度 
6. qua_cntu: 濁度(NTU) 
7. qua_cl: 餘氯(mg/L) 
8. qua_ph: 酸度(pH) 
9. 偵測值-9係維修中，暫停偵測。

## import essential libraries

```python
import urllib # for loading webpage
from bs4 import BeautifulSoup # for parsing webpage
```

## Load and parse webpage

```python
# Loading webpage
response = urllib.urlopen("http://waterpark.twd.gov.tw/www/opendata/wqb/wqb.asmx/GetQualityData").read()

# Parsing webpage
soup = BeautifulSoup(response)
lines = soup.findAll(text=True) 
```

## Processing lines data (string separated by '\n')

```python
data_list = []
for line in lines:
  if line != u'\n': # if not equal to blank line
     data_list.append(line.strip())  
```

```python
for d in data_list:
    print d
```

## Fold 1D list to 2D according to a length = 9

```python
tp_water = [list(l) for l in zip(*[iter(data_list)]*9)]
for loc in tp_water:
    print loc[0], loc[1], loc[2], loc[3], loc[4], loc[7], loc[8]
```

## Convert list to dictionary
* {id:{'time':[date, time], 'name':name, 'location':[lng, lat], 'index':[cntu, cl, ph]}}

```python
water_dict = {}
for d in tp_water:
    water_dict.setdefault(d[2], {'time':[d[0], d[1]], 
                                 'name':d[3], 
                                 'location':[float(d[4]), float(d[5])], 
                                 'index':[float(d[6]), float(d[7]), float(d[8])]})
```
